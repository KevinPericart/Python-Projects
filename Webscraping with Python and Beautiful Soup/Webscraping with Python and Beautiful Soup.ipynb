# <center> Webscraping with Python and Beautiful Soup <center>

In this project, we will try to obtain the list of available jobs related to Compliance in Luxembourg. 

For this purpose, we will use the Job website [Monster](https://www.monster.lu/fr/).

```python
import requests
from bs4 import BeautifulSoup
url = "https://www.monster.lu/fr/emploi/recherche/?q=Compliance&where=Luxembourg"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
results = soup.find(id = 'ResultsContainer')
print(results.prettify())
```

```python
job_elems = results.find_all('section', class_ = 'card-content')
for job_elem in job_elems : 
    print(job_elem, end = '\n' * 2)
```

```python
for job_elem in job_elems : 
    title_elem = job_elem.find('h2', class_ = 'title')
    company_elem = job_elem.find('div', class_ = 'company')
    location_elem = job_elem.find('div', class_ = 'location')
    print(title_elem)
    print(company_elem)
    print(location_elem)
    print()
```

```python
for job_elem in job_elems : 
    title_elem = job_elem.find('h2', class_ = 'title')
    company_elem = job_elem.find('div', class_ = 'company')
    location_elem = job_elem.find('div', class_ = 'location')
    if None in (title_elem, company_elem, location_elem):
        continue
    print(title_elem.text.strip())
    print(company_elem.text.strip())
    print(location_elem.text.strip())
    print()
```

```python
cpl_jobs = results.find_all('h2', string = lambda text: 'compliance' in text.lower())
print(len(cpl_jobs))
```

```python
for cpl_job in cpl_jobs :
    link = cpl_job.find('a')['href']
    print(cpl_job.text.strip())
    print(f"Apply here : {link}\n")
```
